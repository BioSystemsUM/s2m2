### Packaging

When you write several commands in a file to be executed as a program, this file is called a script. When a program is imported, it is called a module.

As a developer starts implementing several modules, they tend to organize them into what is called a package. So, basically, a package is a group of similar modules that usually have a similar purpose.

Fortunatelly, one can use packages already implemented. Commonly, these packages are available in different repositories (https://pypi.org/, https://anaconda.org/anaconda/repo). To install them, we can use pip or conda, the latter only if anaconda or miniconda is installed (https://docs.conda.io/projects/conda/en/latest/user-guide/install/index.html).

##### Selection

Columns

```python
# Select column "sepal_length" by label
print(df["Series1_NHBE_Mock_1"], '\n') #select column by label (string)
print(df.Series1_NHBE_Mock_1, '\n')    #select column by label (column name)

# Select first column by integer
print(df.iloc[:,0]) #select first column by integer
```

Rows

```python
# Select the first row by integer location
print(df.iloc[0], '\n') 

# Select first 2 rows
print(df[0:2]) #select first 2 rows
```

Selecting data using conditions

```python
# select rows with value > 50 in column "Series1_NHBE_Mock_1"
print(df[df["Series1_NHBE_Mock_1"] > 50],'\n')

# select values > x for all columns
print(df[df > 50])
```

Operation  | Syntax | Result
------------- | ------------- | -------------
Select column by label | df["col"]  or df.col| Series
Select column by integer | df.iloc[:, loc] | Series
Select row by label | df.loc["label"] | Series
Select row by integer location | df.iloc[loc] | Series
Slice rows | df[0:2] | DataFrame
Slice columns | df.iloc[:, 0:2] | DataFrame
Slice rows and columns | df.iloc[0:2, 0:2] | DataFrame
Get cell value by label | df.loc["label", "col"] | Scalar value
Get cell value by integer location | df.iloc[2, 0] | Scalar value
Select rows by boolean vector | df["bool_vec"] | DataFrame
